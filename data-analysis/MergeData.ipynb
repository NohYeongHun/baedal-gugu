### 01. weather_quantity.csv + InstagramFoodResult.csv Ìï©Î≥ë

```python
import pandas as pd
weatherDf = pd.read_csv('weather_quantity.csv')
foodDf = pd.read_csv('InstagramFoodResult.csv')
```

```python
mergedDf = pd.merge(left = weatherDf , right = foodDf, how = "outer", on = "date")
mergedDf
```

```python
mergedDf['weekdayorend'] = mergedDf['day_x'].apply(lambda x: 'weekend' if x in ['Saturday','Sunday'] else float("nan") if pd.isna(x) else 'weekday')
```

```python
# Ìï©ÏπòÍ∏∞
weatherDayEndFoodDF = mergedDf.dropna().groupby(['weather','weekdayorend'])['korean','western','dessert','porkcutlets','japanese','chicken','bunsik','chinese','pizza','packedmeal','sashimi'].sum()
```

```python
#ÎπÑÏò§ÎäîÎÇ† ÌèâÏùº ÏùåÏãù ÏàúÏúÑ
weatherDayEndFoodDF.T['ÎπÑ']['weekday'].sort_values(ascending=False)
```

```python
#ÎπÑÏò§ÎäîÎÇ† Ï£ºÎßê ÏùåÏãù ÏàúÏúÑ
weatherDayEndFoodDF.T['ÎπÑ']['weekend'].sort_values(ascending=False)
```
