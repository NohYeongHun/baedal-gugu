### 02. 요일별 태그 데이터

#### 해당 태그 많은 요일 순서: 의미 없음 (데이터 수집 목적: 주문량 X, 카테고리 O)

```python
df_i.groupby(['day']).sum()['cnt'].sort_values(ascending=False)
```

#### 월요일 인기 음식 카테고리

```python
#year, month hour 컬럼 추가
df_i=df_i.reset_index()
df_i['year'] = df_i['date'].apply(lambda x: x.split("-")[0])
df_i['month'] = df_i['date'].apply(lambda x: x.split("-")[1])
df_i['hour'] = df_i['date'].apply(lambda x: x.split(" ")[1])
df_i['hour'] = df_i['hour'].apply(lambda x: x.split(":")[0])
df_i
```
