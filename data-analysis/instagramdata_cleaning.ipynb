```python
data7 = pd.read_csv('baedal.csv', header=None)
data8 = pd.read_csv('baedal2.csv', header=None)
data9 = pd.read_csv('baedal3.csv', header=None) 
data10 = pd.read_csv('baedal4.csv', header=None)
data11 = pd.read_csv('baedal5.csv', header=None)
data12 = pd.read_csv('baedal6.csv', header=None)
data13 = pd.read_csv('baedal7.csv', header=None)
data14 = pd.read_csv('baedal8.csv', header=None)
data15 = pd.read_csv('baedal9.csv', header=None)
data16 = pd.read_csv('baedal10.csv', header=None)
data17 = pd.read_csv('baedal11.csv', header=None)
```

```python
data7['key'] = data7[1]
data7['date'] = data7[2]
data7['tags'] = data7[6]

data8['key'] = data8[1]
data8['date'] = data8[2]
data8['tags'] = data8[6]

data9['key'] = data9[1]
data9['date'] = data9[2]
data9['tags'] = data9[6]

data10['key'] = data10[1]
data10['date'] = data10[2]
data10['tags'] = data10[6]

data11['key'] = data11[1]
data11['date'] = data11[2]
data11['tags'] = data11[6]

data12['key'] = data12[1]
data12['date'] = data12[2]
data12['tags'] = data12[6]

data13['key'] = data13[1]
data13['date'] = data13[2]
data13['tags'] = data13[6]

data14['key'] = data14[1]
data14['date'] = data14[2]
data14['tags'] = data14[6]

data15['key'] = data15[1]
data15['date'] = data15[2]
data15['tags'] = data15[6]

data16['key'] = data16[1]
data16['date'] = data16[2]
data16['tags'] = data16[6]

data17['key'] = data17[1]
data17['date'] = data17[2]
data17['tags'] = data17[6]

df5 = pd.concat([data7[['key','date','tags']], data8[['key','date','tags']], data9[['key','date','tags']], data10[['key','date','tags']], data11[['key','date','tags']], data12[['key','date','tags']], data13[['key','date','tags']], data14[['key','date','tags']], data15[['key','date','tags']], data16[['key','date','tags']], data17[['key','date','tags']]])
df5
```

```python
# 중복 제거하고 데이터 합치기
sumDf = pd.concat([df1,df2,df3,df4,df5]).drop_duplicates(['key'], keep='first') #얘까지 수정하면 돼

# 빈 tags 제거
sumDf = sumDf[sumDf.tags != "[]"]
sumDf = sumDf.dropna()

# date 값 시간만 남기고 실제 Timestamp 형식으로 변경
sumDf['date'] = sumDf['date'].apply(lambda x: pd.to_datetime(x.split(':')[0] + ":00"))

# tags 값 list 형식으로 변경
sumDf['tags'] = sumDf.tags.str[1:-1].str.split(',').tolist()

# date, tags 컬럼만 남김
sumDf = sumDf[['date','tags']]

# 시간별 tags 합치고 2차원 리스트 풀기
sumDf = sumDf.groupby('date').agg({"tags": lambda x: list(x)})
sumDf['tags'] = sumDf['tags'].apply(lambda x: [item for sublist in x for item in sublist])

# 요일 추가
#sumDf['day'] = sumDf['date'].dt.day_name()

# index 초기화
sumDf = sumDf.reset_index()
sumDf
```

```python
#한식 korean
#양식 western
#디저트 dessert
#돈까스 porkcutlets
#일식 japanese
#치킨 chicken
#분식 bunsik
#중국집 chinese
#피자 pizza
#도시락 packedmeal
#회 sashimi

def foodCnt(tags):
    foodDict = {'korean':0,'western':0,'dessert':0,'porkcutlets':0,'japanese':0,'chicken':0,'bunsik':0,'chinese':0,'pizza':0,'packedmeal':0,'sashimi':0}
    for tag in tags:
        #한식
        if "한식" in tag:
            foodDict['korean']+=1
        elif "국밥" in tag:
            foodDict['korean']+=1
        elif "갈비탕" in tag:
            foodDict['korean']+=1
        elif "설렁탕" in tag:
            foodDict['korean']+=1
        elif "칼국수" in tag:
            foodDict['korean']+=1
        elif "잔치국수" in tag:
            foodDict['korean']+=1
        elif "찌개" in tag:
            foodDict['korean']+=1
        elif "구이" in tag:
            foodDict['korean']+=1
        elif "조림" in tag:
            foodDict['korean']+=1
        elif "삼겹살" in tag:
            foodDict['korean']+=1
        elif "불고기" in tag:
            foodDict['korean']+=1
        elif "#전" in tag:
            foodDict['korean']+=1
        #양식
        elif "양식" in tag:
            foodDict['western']+=1
        elif "파스타" in tag:
            foodDict['western']+=1
        elif "스테이크" in tag:
            foodDict['western']+=1
        elif "감바스" in tag:
            foodDict['western']+=1
        elif "샐러드" in tag:
            foodDict['western']+=1
        elif "바베큐" in tag:
            foodDict['western']+=1
        elif "리조또" in tag:
            foodDict['western']+=1
        #디저트
        elif "디저트" in tag:
            foodDict['dessert']+=1
        elif "케이크" in tag:
            foodDict['dessert']+=1
        elif "케익" in tag:
            foodDict['dessert']+=1
        elif "마카롱" in tag:
            foodDict['dessert']+=1
        elif "와플" in tag:
            foodDict['dessert']+=1
        elif "빙수" in tag:
            foodDict['dessert']+=1
        elif "설빙" in tag:
            foodDict['dessert']+=1
        elif "아이스크림" in tag:
            foodDict['dessert']+=1
        elif "베라" in tag:
            foodDict['dessert']+=1
        elif "베스킨라빈스" in tag:
            foodDict['dessert']+=1
        elif "크로플" in tag:
            foodDict['dessert']+=1
        elif "붕어빵" in tag:
            foodDict['dessert']+=1
        elif "호떡" in tag:
            foodDict['dessert']+=1
        #돈까스
        elif "돈까스" in tag:
            foodDict['porkcutlets']+=1
        elif "돈가스" in tag:
            foodDict['porkcutlets']+=1
        elif "돈가츠" in tag:
            foodDict['porkcutlets']+=1
        #일식
        elif "일식" in tag:
            foodDict['japanese']+=1
        elif "스시" in tag:
            foodDict['japanese']+=1
        elif "초밥" in tag:
            foodDict['japanese']+=1
        elif "라멘" in tag:
            foodDict['japanese']+=1
        elif "소바" in tag:
            foodDict['japanese']+=1
        elif "메밀" in tag:
            foodDict['japanese']+=1
        elif "우동" in tag:
            foodDict['japanese']+=1
        elif "덴푸라" in tag:
            foodDict['japanese']+=1
        elif "규동" in tag:
            foodDict['japanese']+=1
        #치킨
        elif "치킨" in tag:
            foodDict['chicken']+=1
        elif "치느님" in tag:
            foodDict['chicken']+=1
        #분식
        elif "분식" in tag:
            foodDict['bunsik']+=1
        elif "떡볶이" in tag:
            foodDict['bunsik']+=1
        elif "순대" in tag:
            foodDict['bunsik']+=1
        elif "튀김" in tag:
            foodDict['bunsik']+=1
        elif "튀김" in tag:
            foodDict['bunsik']+=1
        elif "김밥" in tag:
            foodDict['bunsik']+=1
        #중식
        elif "중식" in tag:
            foodDict['chinese']+=1
        elif "짜장면" in tag:
            foodDict['chinese']+=1
        elif "자장면" in tag:
            foodDict['chinese']+=1
        elif "짜장" in tag:
            foodDict['chinese']+=1
        elif "짬뽕" in tag:
            foodDict['chinese']+=1
        elif "탕수육" in tag:
            foodDict['chinese']+=1
        elif "깐풍기" in tag:
            foodDict['chinese']+=1
        elif "깐쇼새우" in tag:
            foodDict['chinese']+=1
        elif "마라탕" in tag:
            foodDict['chinese']+=1
        elif "마라샹궈" in tag:
            foodDict['chinese']+=1
        elif "꿔바로우" in tag:
            foodDict['chinese']+=1
        #피자
        elif "피자" in tag:
            foodDict['pizza']+=1
        #도시락
        elif "도시락" in tag:
            foodDict['packedmeal']+=1
        #회
        elif "#회" in tag:
            foodDict['sashimi']+=1
        elif "광어" in tag:
            foodDict['sashimi']+=1
        elif "우럭" in tag:
            foodDict['sashimi']+=1
        elif "연어" in tag:
            foodDict['sashimi']+=1
        elif "참치" in tag:
            foodDict['sashimi']+=1
        elif "방어" in tag:
            foodDict['sashimi']+=1
    return foodDict
```

```python
# 음식 딕셔너리 컬럼 생성
sumDf['tagsDict'] = sumDf['tags'].apply(lambda x: foodCnt(x))
```

```python
# 음식 종류별 컬럼 생성
sumDf['korean'] = sumDf['tagsDict'].apply(lambda x: x['korean'])
sumDf['western'] = sumDf['tagsDict'].apply(lambda x: x['western'])
sumDf['dessert'] = sumDf['tagsDict'].apply(lambda x: x['dessert'])
sumDf['porkcutlets'] = sumDf['tagsDict'].apply(lambda x: x['porkcutlets'])
sumDf['japanese'] = sumDf['tagsDict'].apply(lambda x: x['japanese'])
sumDf['chicken'] = sumDf['tagsDict'].apply(lambda x: x['chicken'])
sumDf['bunsik'] = sumDf['tagsDict'].apply(lambda x: x['bunsik'])
sumDf['chinese'] = sumDf['tagsDict'].apply(lambda x: x['chinese'])
sumDf['pizza'] = sumDf['tagsDict'].apply(lambda x: x['pizza'])
sumDf['packedmeal'] = sumDf['tagsDict'].apply(lambda x: x['packedmeal'])
sumDf['sashimi'] = sumDf['tagsDict'].apply(lambda x: x['sashimi'])
```
