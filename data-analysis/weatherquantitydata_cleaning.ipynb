```python
def checkWeather(num):
    if num == 0:
        return '맑음'
    elif num == 2 or num == 3 or num == 6 or num == 7:
        return '눈'
    else:
        return '비'
```

```python
def linkDfData(df, data):
    dataList = list(data.iloc[244868].items())

    newDf = pd.DataFrame({})
    for i in range(len(dataList)):
        if i == 0:
            date = dataList[0][0]
        else:
            newDf = pd.concat([newDf,pd.DataFrame({'date':pd.to_datetime(date+' ' +dataList[i][0].split('h')[0]+':00'),'weather' : checkWeather(dataList[i][1])}, index=[i])])

    return pd.concat([df, newDf])
```

```python
import os
path = 'C:\workspace\elice_jupyter\data\weather'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.csv')] ## 파일명 끝이 .csv인 경우
file_list_py
```

```python
df = pd.DataFrame({})
for filename in file_list_py:
    data = pd.read_csv(f'C:\workspace\elice_jupyter\data\weather\{filename}', skiprows=23, on_bad_lines='skip')
    df = linkDfData(df, data)
df.to_csv('sum.csv')
```
